```python
# Imports
import tensorflow as tf
import numpy as np
import random
from random import randint
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Activation
from tensorflow.keras import losses
import tensorflow.keras.backend as backend

from sklearn.preprocessing import minmax_scale

import matplotlib.pyplot as plt
```

```python
# Define the data
stimulusResponsePairCount = 10
dataReplicationCount = 1
```

```python
# Create x and y
x = []
y = []
for i in range(stimulusResponsePairCount):
  a = []
  b = []
  for j in range(1):
    a.append(i)
    b.append(randint(0, 255))
  x.append(a)
  y.append(b)

xOrig = x.copy()
yOrig = y.copy()
for i in range(dataReplicationCount):
  xi = xOrig.copy()
  x.extend(xi)
  yi = yOrig.copy()
  y.extend(yi)

x = np.asarray(x)
y = np.asarray(y)
```

```python
x = minmax_scale(x, feature_range=(0,1))
y = minmax_scale(y, feature_range=(0,1))
```

```python
# Define the model

model = Sequential()
model.add(Dense(32, activation='relu', input_dim=1))
model.add(Dense(32, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy', 'mean_squared_error'])

model.summary()
```

```python
# Fit the model

history = model.fit(x, y, epochs=3000, batch_size=1000)
```

```python
# Output the result

misses = 0
for i in range(len(x)):
  xi = x[i]
  yi = y[i]
  pred = model.predict(xi)
  print('x = ' + str(xi) + ' y = ' + str(yi) + ' pred = ' + str(pred))
  if yi[0] != pred:
    misses = misses + 1
print('Misses = ' + str(misses) + ' out of ' + str(stimulusResponsePairCount))
```
